```python
#{2}
#[Demo]와 [구매내역정보] 불러오기

cs = pd.read_csv('Demo.csv', encoding='cp949', engine='python')
tr = pd.read_csv('구매내역정보.csv', encoding='cp949', engine='python')
```

```python
#{4}
#성별 카테고리를 1,2에서 남,여 로

cs.성별 = cs.성별.map({1:'남', 2:'여'})
```

```python
#{5}
#'거주지역' 세분화

cs['거주지역_광역'] = cs.거주지역.apply(lambda x: x.split()[0])
cs['거주지역_기초'] = cs.거주지역.apply(lambda x: x.split()[1])
```

```python
#{6}
#'구매지역' 세분화

tr['구매지역_광역']=tr.구매지역.apply(lambda x: x.split()[0])
tr['구매지역_기초']=tr.구매지역.apply(lambda x: x.split()[1])
```

```python
#{7}
#[Demo]와 [구매내역정보]를 merge

md = pd.merge(cs, tr, on='ID')
md
```

```python
#{8}
#65세 이상 소비자

md65 = md.query('65<=연령')
md65
```

```python
#{9}
#서울,경기에 거주하며 서울 지점에서 구매하는 65세 이상 여성

n = md.query('65<=연령 and 성별=="여" and (거주지역_광역=="서울" or 거주지역_광역=="경기") and 구매지역_광역=="서울"')
n
```

```python
#{10}
#'월' column만들기

n['월'] = n.구매일자.astype(str).str[4:6]
pd.pivot_table(n, index = 'ID', columns = '월', values = '구매금액', aggfunc = sum, fill_value=0).reset_index()
```

```python
#{11}
#'요일' column 만들기

n['요일'] = pd.to_datetime(n.구매일자.astype(str)).dt.weekday_name
pd.pivot_table(n, index='성별', columns='요일', values='구매금액', aggfunc=sum, fill_value=0).reset_index()

def fw(x) :
    return(('월', '화', '수', '목', '금', '토', '일')[x.weekday()])

n['요일'] = pd.to_datetime(n.구매일자.astype(str)).apply(fw) + '요일'
n['요일'] = n.요일.astype('category').cat.reorder_categories(['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일'], ordered = True)
n
```

```python
#{12}
#데이터 형식을 문자로

n.구매일자 = n.구매일자.apply(str)
n.구매시간 = n.구매시간.apply(str)
```

```python
#{13}
#새롭게 상품분류 '소분류1'

def f(x):
    if x in['농산물','수산물']:
        return '농수산물'
    elif x in ['여성용의류-정장류','남성용의류-정장류']:
        return '정장류'
    elif x in ['여성용의류-캐주얼','여성용의류-SPA', '남성용의류-캐주얼']:
        return '캐주얼'
    elif x in ['패션잡화','주얼리']:
        return '패션잡화/주얼리'
    elif x in ['PC/노트북/프린터/카메라','휴대폰/태블릿','컴퓨터주변기기']:
        return '전자기기'
    elif x in ['스포츠용품/의류','골프용품/의류']:
        return '스포츠+골프용품/의류'
    else:
        return x
    
n['소분류1'] = n.상품중분류명.apply(f)
n
```

```python
#{14}
#구매금액 분류

def r(x):
    if 0 <= x < 10000:
        return '0-1만원'
    elif 10000 <= x <20000:
        return '1-2만원'
    elif 20000 <= x <30000:
        return '2-3만원'
    elif 30000 <= x < 40000:
        return '3-4만원'
    elif 40000 <= x < 50000:
        return '4-5만원'
    elif 50000 <= x < 70000:
        return '5-7만원'
    elif 70000 <= x < 90000:
        return '7-9만원'
    elif 90000 <= x < 110000:
        return '9-11만원'
    elif 110000 <= x < 200000:
        return '11-20만원'
    elif 200000 <= x < 500000:
        return '20-50만원'
    elif 500000 <= x < 1000000:
        return '50-100만원'
    elif 1000000 <= x < 2000000:
        return '100-200만원'
    elif 2000000 <= x < 4000000:
        return '200-400만원'
    elif 4000000 <= x < 7000000:
        return '400-700만원'
    
n['구매금액분류'] = n.구매금액.apply(r)
n
```

```python
#{15}
#2014년 날씨 데이터 불러오기
#[날씨]의 기상 관측 지점 숫자를 지역구 이름으로

날씨 = pd.read_csv('구별시간별온도.csv', encoding='cp949', engine='python')
날씨.지점 = 날씨.지점.map({400:'강남구', 402:'강동구', 424:'강북구', 404:'강서구', 509:'관악구', 413:'광진구', 423:'구로구', 417:'금천구', 425:'남현구', 407:'노원구', 406:'도봉구', 408:'동대문구', 411:'마포구', 422:'동작구', 412:'서대문구', 401:'서초구', 421:'성동구', 414:'성동구', 403:'송파구', 405:'양천구', 510:'영등포구', 415:'용산구', 416:'은평구', 419:'중구',409:'중랑구', 889:'동작구'})
날씨['구매지역_기초']=날씨.지점
날씨
```

```python
#{16}
#[날씨]에 '구매일자'와 '구매시간' column 만들기

날씨['구매일자'] = 날씨.일시.apply(lambda x: x.split()[0])
날씨['구매시간'] = 날씨.일시.apply(lambda x: x.split()[1])
```

```python
#{17}
#[날씨]의 구매일자와 구매시간을 [n]의 구매일자와 구매시간 형식으로 맞추기

날씨['구매일자'] = 날씨.구매일자.apply(lambda y: y.split(':')[0])
def s(x):
    if '-' in x:
        return x.replace("-","")
    else:
        return x

날씨['구매일자'] = 날씨.구매일자.apply(s)

날씨['구매시간'] = 날씨.구매시간.apply(lambda y: y.split(':')[0])
def s(x):
    if x.startswith("0"):
        return x.replace("0","")
    else:
        return x

날씨['구매시간'] = 날씨.구매시간.apply(s)

날씨
```

```python
#{18}
#[날씨]와 [n] 을  '구매지역_기초, 구매일자, 구매시간'을 기준으로 merge

nn = pd.merge(n, 날씨, on=['구매지역_기초','구매일자','구매시간'])
nn
```

```python
#{19}
#column 이름 변경
#기온(°C) -> 기온
#강수량(mm) -> 강수량

nn = nn.rename(columns = {'기온(°C)': '기온'})
nn = nn.rename(columns = {'강수량(mm)': '강수량'})
nn
```

```python
#{20}
#불필요한 column 제거

nn.drop(['거주지역','구매지역','구매지역_광역','지점','강수량','일시','습도(%)'], axis=1)
```

```python
#{21}
#기온 분류

def t(x):
    if x < -10.0 :
        return '-10.0 이하'
    elif -10.0 <= x < -5.0:
        return '-10.0~-0.5'
    elif -5.0 <= x < 0:
        return '-5.0~0'
    elif 0 <= x < 5:
        return '0~5'
    elif 5 <= x < 10:
        return '5~10'
    elif 10 <= x < 15:
        return '10~15'
    elif 15 <= x < 20:
        return '15~20'
    elif 20 <= x < 25:
        return '20~25'
    elif 25 <= x < 30:
        return '25~30'
    elif 30 <= x < 35:
        return '30~35'
    
nn['기온분류'] = nn.기온.apply(t)
nn
```
